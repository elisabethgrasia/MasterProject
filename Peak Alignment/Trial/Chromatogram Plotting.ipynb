```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import savgol_filter
np.set_printoptions(precision=2)
```

```python
data = pd.read_csv('peak alignment_csv data testing.csv')
data.head()
```

```python
X, y = data.iloc[:, 0].values, data.iloc[:, 1:].values

plt.plot(X, y)
plt.xlabel('Mass')
plt.show()
```

```python
X, y = data.iloc[0:316, 0], data.iloc[1, 1:]

plt.plot(X, y)
plt.ylabel('Spectra')
plt.xlabel('Mass')
plt.show()
```

```python
Xsg = savgol_filter(X, 10, 2)

plt.plot(Xsg, y)
plt.ylabel('Spectra')
plt.xlabel('Mass')
plt.show()
```
